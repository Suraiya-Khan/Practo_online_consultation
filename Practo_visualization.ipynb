```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Practo.csv')
```

```python
print("\n--- 1. Doctor's Experience Analysis ---")
df['Experience'].describe()
```

```python
df.describe()
```

```python
# Computing averages grouped by Department and City
avg_grouped = df.groupby(['City','Department' ])[['Experience', 'Consultation Fee']].mean().reset_index()
avg_grouped.columns = ['City','Department', 
                   'Avg_Experience', 'Avg_Consultation_Fee']

# Round the averages for readability
avg_grouped[['Avg_Experience', 'Avg_Consultation_Fee']] = \
    avg_grouped[['Avg_Experience', 'Avg_Consultation_Fee']].round(2)

avg_grouped
```

```python
avg_grouped
```

```python
max_grouped = df.groupby(['City','Department' ])[['Experience', 'Consultation Fee', 'Ratings', 'Patient Stories']].max().reset_index()
max_grouped.columns = ['City','Department', 
                   'max_Experience', 'max_Consultation_Fee', 'max_Ratings', 'max_Patient_Stories']

max_grouped[['max_Experience', 'max_Consultation_Fee', 'max_Ratings', 'max_Patient_Stories']] = \
    max_grouped[['max_Experience', 'max_Consultation_Fee', 'max_Ratings', 'max_Patient_Stories']].round(2)

max_grouped
```

```python
min_grouped = df.groupby(['City','Department' ])[['Experience', 'Consultation Fee', 'Ratings', 'Patient Stories']].min().reset_index()
min_grouped.columns = ['City','Department', 
                   'min_Experience', 'min_Consultation_Fee', 'min_Ratings', 'min_Patient_Stories']

min_grouped[['min_Experience', 'min_Consultation_Fee', 'min_Ratings', 'min_Patient_Stories']] = \
    min_grouped[['min_Experience', 'min_Consultation_Fee', 'min_Ratings', 'min_Patient_Stories']].round(2)

min_grouped
```

```python
# Visualization of experience
plt.figure(figsize=(12,5))
sns.histplot(df['Experience'],bins=15,kde=True)
plt.title('Distribution of Doctor Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Number of Doctors')
plt.grid(axis = 'y',alpha=0.75)
plt.show()
```

```python

# Define bin edges including 20–30 and 30–40
bins = [0, 5, 10, 15, 20, 30, 40, df['Experience'].max()]
# Calculate number of bins; must equal len(bins) - 1
num_bins = len(bins) - 1

# Define the appropriate number of labels
labels = [
    '0–5 yrs', '6–10 yrs', '11–15 yrs',
    '16–20 yrs', '21–30 yrs', '31–40 yrs', f'41–{int(df['Experience'].max())} yrs'
]
assert len(labels) == num_bins, f"Need {num_bins} labels, but got {len(labels)}."

# Categorize using pd.cut
exp_categories = pd.cut(
    df['Experience'],
    bins=bins,
    labels=labels,
    include_lowest=True,
    right=False
)

# Generate counts and plot
counts = exp_categories.value_counts().reindex(labels)

plt.figure(figsize=(6, 8))
counts.plot(
    kind='pie',
    labels=counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=plt.cm.tab20.colors
)
plt.title('Doctor Experience Distribution')
plt.ylabel('')
plt.axis('equal')
plt.legend(title='Experience Ranges', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
Avg_ratings= df['Ratings'].mean()
print('Average ratings of doctors :',Avg_ratings)
max_rating = df['Ratings'].max()
print('Maximum ratings:',max_rating)
min_rating = df['Ratings'].min()
print('Minimum rating:',min_rating)
```

```python
# Visulaization ratings

bins = [0, 60, 70, 80, 90, 100]
labels = ['0–60%', '61–70%', '71–80%', '81–90%', '91–100%']
rating_bins = pd.cut(df['Ratings'], bins=bins, labels=labels, include_lowest=True)

counts = rating_bins.value_counts().reindex(labels)

plt.figure(figsize=(10, 6))
sns.barplot(x=counts.index, y=counts.values, palette='bright')
plt.title('Rating Range Distribution')
plt.xlabel('Rating Range')
plt.ylabel('Number of Doctors')
plt.xticks(rotation=45)
plt.show()
```

```python
print("\n--- 1. Patients stories Analysis ---")
df['Patient Stories'].describe()
```

```python
# Stories Visulaization

bins = [0, 50, 100, 200, 500, df['Patient Stories'].max()]
labels = ['0–50', '51–100', '101–200', '201–500', f'501–{int(df["Patient Stories"].max())}']

df['story_bins'] = pd.cut(df['Patient Stories'], bins=bins, labels=labels, include_lowest=True)

# Count per bin
grouped = df['story_bins'].value_counts().sort_index().reset_index()
grouped.columns = ['Story Range', 'Doctor Count']

# Plot
sns.set_theme(style="whitegrid")
sns.barplot(x='Story Range', y='Doctor Count', data=grouped, palette='viridis')
plt.title('Distribution of Patient Stories by Range')
plt.xlabel('Patient Stories Range')
plt.ylabel('Number of Doctors')
plt.xticks(rotation=45)
plt.show()
```

```python
# city visualization
# Compute counts per city
city_counts = df['City'].value_counts().sort_values(ascending=False)

# Get the city with the maximum doctors
top_city = city_counts.idxmax()
top_count = city_counts.max()

print(f"City with the most doctors: {top_city} ({top_count} doctors)")

# Visualize using a bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=city_counts.index, y=city_counts.values, palette='magma')
plt.title('Number of Doctors by City')
plt.xlabel('City')
plt.ylabel('Number of Doctors')
plt.xticks(rotation=45, ha='right')
plt.axhline(y=top_count, color='red', linestyle='--', 
            label=f'Top City: {top_city}')
plt.legend()
plt.tight_layout()
plt.show()
```

Observed Pattern and Possible Interpretation
1. Higher fee → higher rating- Perceived quality aligns with price; patients may equate cost with value.
2.Flat/no trend	Ratings are independent of fee—other factors drive satisfaction.
3.Higher fee → lower rating	Potential value misalignment—patients may feel overcharged.
4.Wide scatter	More complex influences at work—investigate by department, city, etc.
